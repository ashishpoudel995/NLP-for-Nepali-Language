```python
import math
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

Original Text to be summarized

```python
text='नेपालमा प्याजको मूल्य भारतीय बजारसँग निर्भर छ, उता बढे यता पनि बढ्छ । भारतबाट दैनिक ८० देखि १०० टनसम्म प्याज आयात हुँदै आएको छ । कुनै समय आयात कम हुँदा मूल्य बढ्ने गरेको थियो । हाम्रो बजारमा खपत कम भए पनि भारतीय बजारका कारण प्याजको मूल्य प्रतिकिलो ८० रुपैयाँ थोकमै पर्न थालेको छ । आयात कम भएर यसो भएको होइन । भारतमै प्याजको मूल्य एक सातामा प्रतिकिलो २० रुपैयाँसम्म बढेको पाइएको छ । यसले गर्दा यता पनि प्याजको मूल्य लगातार बढिरहेको छ । नयाँ उत्पादन आउन समय लाग्ने हुनाले अबको केही समय प्याजको मूल्य उकालो लाग्ने सम्भावना छ ।हाल बजारमा भइरहेको प्याज सबै आयातित हो । स्वदेशी उत्पादनले थोरै समयका लागि मात्र माग धान्ने गरेको छ । मागपूर्तिका लागि बर्सेनि २ अर्बभन्दा बढी रुपैयाँको प्याज भारतबाट आयात हुने गरेको छ । सबैजसो माग भारतले धान्ने गरेको छ । भारतीय बजारमै प्याजको माग धेरै हुने गरेको छ । उत्पादन कम र माग धेरै भएकाले केही वर्षदेखि प्याजको मूल्यमा उतारचढाव आइरहेको छ । उताबाट मूल्य घटेर आए मात्र स्थानीय बजारमा प्याजको मूल्य घट्ने हो ।नेपालले प्याजकै आयातमा पनि अर्बौं खर्च गर्ने गरेको छ । प्याज भारतबाट मात्र होइन, केही मात्रामा चीनबाट पनि आयात हुने गरेको छ । भन्सार विभागको तथ्यांकअनुसार, आर्थिक वर्ष सन् २०७६–७७ मा ४.२१ अर्ब रुपैयाँ बराबरको १ लाख ७ हजार टन प्याज आयात गरेको थियो । त्यो पनि कोभिड महामारीका कारण घटेको हो । योभन्दा अघिल्लो वर्ष २०७५–७६ मा १ लाख ७८ हजार टन प्याज आयात गर्न ५ अर्ब ६२ करोड बिदेसिएको थियो । अर्कातिर, तीन वर्षयता स्थानीय उत्पादनमा लगातार वृद्धि भइरहेको छ । नेपालमा बर्सेनि करिब ३ लाख टन प्याज उत्पादन हुन्छ, जबकि खपत करिब ४ लाख टन हुने गरेको छ ।'
print(text)
len(text)
```

Splitting all the sentences of the original text

```python
documents=re.split('।',text)
```

```python
documents_size = len(documents)
print(documents_size)
```

Splitting all the words of the original text

```python
words=text.split()
print(words)
```

Creating frequency matrix to keep the count of each unique words in each sentence

```python
def create_freq_matrix(documents):
    frequency_matrix = {}
    stopWords = set(stopwords.words("nepali"))

    for i in range(len(documents)):
        freq_table = {}
        words=documents[i].split()
        for word in words:
            if word in stopWords:
                continue

            if word in freq_table:
                freq_table[word] += 1
            else:
                freq_table[word] = 1

        frequency_matrix[i] = freq_table

    return frequency_matrix

freq_matrix = create_freq_matrix(documents)
freq_matrix
```

Counting term frequency for each word and storing in term frequency matrix

```python
def tf(freq_matrix):
    tf_matrix = {}

    for sent, f_table in freq_matrix.items():
        tf_table = {}

        count_words_in_sentence = len(f_table)
        for word, count in f_table.items():
            tf_table[word] = count / count_words_in_sentence

        tf_matrix[sent] = tf_table

    return tf_matrix

tf_matrix = tf(freq_matrix)
tf_matrix
```

Calculating Inverse Document Frequency for each word

```python
def idf(freq_matrix, doc_per_words, documents_size):
    idf_matrix = {}

    for sent, f_table in freq_matrix.items():
        idf_table = {}

        for word in f_table.keys():
            idf_table[word] = math.log10(documents_size / float(doc_per_words[word]))

        idf_matrix[sent] = idf_table

    return idf_matrix
def documents_per_words(freq_matrix):
    word_per_doc_table = {}

    for sent, f_table in freq_matrix.items():
        for word, count in f_table.items():
            if word in word_per_doc_table:
                word_per_doc_table[word] += 1
            else:
                word_per_doc_table[word] = 1

    return word_per_doc_table
doc_per_words = documents_per_words(freq_matrix)
```

Storing the IDF values of all words in the IDF matrix

```python
idf_matrix = idf(freq_matrix,doc_per_words, documents_size)
idf_matrix
```

Calculating TF-IDF values of each word in the corpus

```python
def tf_idf(tf_matrix, idf_matrix):
    tf_idf_matrix = {}

    for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), idf_matrix.items()):

        tf_idf_table = {}

        for (word, val1), (w, val2) in zip(f_table1.items(),f_table2.items()):  
            tf_idf_table[word] = float(val1 * val2)

        tf_idf_matrix[sent1] = tf_idf_table

    return tf_idf_matrix
tf_idf_matrix = tf_idf(tf_matrix, idf_matrix)
tf_idf_matrix
```

Providing specific weight to each sentence (Weight of a sentence= total weight/number of words)

```python
def sentence_weights(tf_idf_matrix) -> dict:
    

    sentenceValue = {}

    for sent, f_table in tf_idf_matrix.items():
        total_weight_per_sentence = 0

        count_words_in_sentence = len(f_table)
        for word, weight in f_table.items():
            total_weight_per_sentence += weight
        if count_words_in_sentence !=0:
            sentenceValue[sent] = total_weight_per_sentence / count_words_in_sentence
        else:
            sentenceValue[sent]=0
    return sentenceValue
weights_sentence = sentence_weights(tf_idf_matrix)
print(weights_sentence)
```

Using average of all weights to calculate the cutoff weight

```python
def find_cutoff(sentenceValue) -> int:
    sumValues = 0
    for entry in sentenceValue:
        sumValues += sentenceValue[entry]

    average = (sumValues / len(sentenceValue))

    return average
cutoff_weight = find_cutoff(weights_sentence)
cutoff_weight
```

Filtering the weights of all the sentences using cutoff weight

```python
def text_summary(docs, weights, cutoff):
    sentence_count = 0
    summary = []

    for i in range(len(docs)):
        if weights[i] >= (cutoff):
            summary.append(docs[i])
            sentence_count += 1

    return summary
```

Summarizing the text

```python
summary = '।'.join(text_summary(documents, weights_sentence,0.9*cutoff_weight ))
print(summary+'।')
```
