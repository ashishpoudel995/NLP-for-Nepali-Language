````python
verb_stems_type1=["घुम्","सुत्","पढ्","पर्","गर्","हेर्","सुन्","भन्","बोल्","चल्","मुत्","लेख्","बुझ्","उठ्","खेल्","सोच्","खोज्","हास्","हाक्","भेट्","बेच्","किन्","देख्","थुक्","चोर्","जीत्","खोल्","पोल्","काट्","सक्","सिक्","रोक्","थाप्","पुग्","पिट्","बग्","चल्","मान्","माग्","नाच्","बाच्","फेर्","मार्","खिच्","थिच्","फुल्","बाल्","निभ्","कुद्","तात्","हार्","राख्","मिल्","खस्"]

conjugated_forms_type1=[]

##All possible conjugations for Regular Verbs- Type 1

for verb in verb_stems_type1:
  #Affirmative Paradigm
  #1s
  conjugated_forms_type1.append(verb+"छु")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"े")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ाैला")

  #1p
  conjugated_forms_type1.append(verb+"छौ")
  conjugated_forms_type1.append(verb+"यौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ाैला")

  #2,s,m,LGH
  conjugated_forms_type1.append(verb+"छस्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िस्")
  conjugated_forms_type1.append(verb+"लास्")

  #2,s,f,LGH
  conjugated_forms_type1.append(verb+"छेस्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िस्")
  conjugated_forms_type1.append(verb+"लिस्")

  #2,s,m,MGH
  conjugated_forms_type1.append(verb+"छौ")
  conjugated_forms_type1.append(verb+"यौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ाैला")

  #2,s,f,MGH
  conjugated_forms_type1.append(verb+"छेउ")
  conjugated_forms_type1.append(verb+"यौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ाैली")

  #2,p,m/f,MGH
  conjugated_forms_type1.append(verb+"छौ")
  conjugated_forms_type1.append(verb+"यौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ाैला")

  #3,s,m,LGH
  conjugated_forms_type1.append(verb+"छ")
  conjugated_forms_type1.append(verb+"यो")
  conjugated_forms_type1.append(verb+"ला")

  #3,s,f,LGH
  conjugated_forms_type1.append(verb+"छे")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ी")
  conjugated_forms_type1.append(verb+"ली")

  #3,s,m,MGH
  conjugated_forms_type1.append(verb+"छन्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"े")
  conjugated_forms_type1.append(verb+"लान्")

  #3,s,f,MGH
  conjugated_forms_type1.append(verb+"छिन्")
  conjugated_forms_type1.append(verb+"िन्")
  conjugated_forms_type1.append(verb+"लिन्")

  #3,p,m/f,MGH
  conjugated_forms_type1.append(verb+"छन्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"े")
  conjugated_forms_type1.append(verb+"लान्")

  #HGH
  conjugated_forms_type1.append(verb+"नुहुन्छ")
  conjugated_forms_type1.append(verb+"नुभयो")
  conjugated_forms_type1.append(verb+"नुहोला")

  #Negative Paradigm
  #1s
  conjugated_forms_type1.append(verb+"दिन")
  conjugated_forms_type1.append(verb+"िन")
  conjugated_forms_type1.append("न"+verb[0:len(verb)-1]+"ुला")
  #1p
  conjugated_forms_type1.append(verb+"दैनौ")
  conjugated_forms_type1.append(verb+"िनौ")
  conjugated_forms_type1.append("न"+verb[0:len(verb)-1]+"ाैला")

  #2,s,m,LGH
  conjugated_forms_type1.append(verb+"दैनस्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िनस्")
  conjugated_forms_type1.append("न"+verb+"लास्")

  #2,s,f,LGH
  conjugated_forms_type1.append(verb+"दैनेस्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िनेस्")
  conjugated_forms_type1.append("न"+verb+"लिस्")

  #2,s,m,MGH
  conjugated_forms_type1.append(verb+"दैनौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ेनौ")
  conjugated_forms_type1.append("न"+verb[0:len(verb)-1]+"ाैला")

  #2,s,f,MGH
  conjugated_forms_type1.append(verb+"दिनौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िनौ")
  conjugated_forms_type1.append("न"+verb[0:len(verb)-1]+"ाैली")

  #2,p,m/f,MGH
  conjugated_forms_type1.append(verb+"दैनौ")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ेनौ")
  conjugated_forms_type1.append("न"+verb[0:len(verb)-1]+"ाैला")

  #3,s,m,LGH
  conjugated_forms_type1.append(verb+"दैन")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ेन")
  conjugated_forms_type1.append("न"+verb+"ला")

  #3,s,f,LGH
  conjugated_forms_type1.append(verb+"दैने")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िन")
  conjugated_forms_type1.append("न"+verb+"ली")

  #3,s,m,MGH
  conjugated_forms_type1.append(verb+"दैनन्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ेनन्")
  conjugated_forms_type1.append("न"+verb+"लान्")

  #3,s,f,MGH
  conjugated_forms_type1.append(verb+"दैनिन्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"िनन्")
  conjugated_forms_type1.append("न"+verb+"लिन्")

  #3,p,m/f,MGH
  conjugated_forms_type1.append(verb+"दैनन्")
  conjugated_forms_type1.append(verb[0:len(verb)-1]+"ेनन्")
  conjugated_forms_type1.append("न"+verb+"लान्")

  #HGH
  conjugated_forms_type1.append(verb+"नुहुदैन")
  conjugated_forms_type1.append(verb+"नुभएन")
  conjugated_forms_type1.append("न"+verb+"नुहोला")

  #Past Habitual Paradigm
  #1s(m/f)
  conjugated_forms_type1.append(verb+"थे")
  conjugated_forms_type1.append(verb+"दिनथे")
  conjugated_forms_type1.append(verb+"दैनथे")

  #1p
  conjugated_forms_type1.append(verb+"थेउ")
  conjugated_forms_type1.append(verb+"दैनथेउ")

  #2,s,m,LGH
  conjugated_forms_type1.append(verb+"थिस्")
  conjugated_forms_type1.append(verb+"दैनथिस्")

  #2,s,f,LGH
  conjugated_forms_type1.append(verb+"थिस्")
  conjugated_forms_type1.append(verb+"दैनथिस्")

  #2,s,m,MGH
  conjugated_forms_type1.append(verb+"थेउ")
  conjugated_forms_type1.append(verb+"दैनथेउ")

  #2,s,f,MGH
  conjugated_forms_type1.append(verb+"थेउ")
  conjugated_forms_type1.append(verb+"दैनथेउ")

  #2,p,m/f,MGH
  conjugated_forms_type1.append(verb+"थेउ")
  conjugated_forms_type1.append(verb+"दैनथेउ")

  #3,s,m,LGH
  conjugated_forms_type1.append(verb+"थ्यो")
  conjugated_forms_type1.append(verb+"दैनथ्यो")

  #3,s,f,LGH
  conjugated_forms_type1.append(verb+"थी")
  conjugated_forms_type1.append(verb+"दैनथी")

  #3,s,m,MGH
  conjugated_forms_type1.append(verb+"थे")
  conjugated_forms_type1.append(verb+"दैनथे")

  #3,s,f,MGH
  conjugated_forms_type1.append(verb+"थिन्")
  conjugated_forms_type1.append(verb+"दैनथिन्")

  #3,p,m/f,MGH
  conjugated_forms_type1.append(verb+"थे")
  conjugated_forms_type1.append(verb+"दैनथे")

  #HGH
  conjugated_forms_type1.append(verb+"नुहुन्थ्यो")
  conjugated_forms_type1.append(verb+"नुहुदैनथ्यो")
  conjugated_forms_type1.append(verb+"नुहुन्नथ्यो")

unique_conjugations_type1=[]
for conjugations in conjugated_forms_type1:
  if conjugations not in unique_conjugations_type1:
    unique_conjugations_type1.append(conjugations)

````
